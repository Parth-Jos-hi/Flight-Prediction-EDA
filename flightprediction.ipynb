```python
#### Feature Engineering process
fp_final['Date'] = fp_final['Date_of_Journey'].str.split('/').str[0]
fp_final['Month'] = fp_final['Date_of_Journey'].str.split('/').str[1]
fp_final['Year'] = fp_final['Date_of_Journey'].str.split('/').str[2]
```

```python
fp_final['Date'] = fp_final['Date'].astype(int)
fp_final['Month'] = fp_final['Month'].astype(int)
fp_final['Year'] = fp_final['Year'].astype(int)
```

```python
fp_final.drop('Date_of_Journey',axis = 1,inplace = True)
```

```python
fp_final.head(1)
```

```python
fp_final['Arrival_Time'] = fp_final['Arrival_Time'].str.split(" ").str[0].head()
```

```python
fp_final.isnull().sum()
```

```python
fp_final['Arrival_hour'] = fp_final['Arrival_Time'].str.split(":").str[0]
fp_final['Arrival_min'] = fp_final['Arrival_Time'].str.split(" ").str[1]
```

```python
fp_final['Arrival_hour'].head()
```

```python
fp_final['Arrival_min'].head()
```

```python
fp_final['Arrival_hour'] = fp_final['Arrival_hour'].astype('Int64')
fp_final['Arrival_min']  = fp_final['Arrival_min'].astype('Int64')
```

```python
fp_final.info()
```

```python
fp_final.drop('Arrival_Time',axis= 1,inplace = True)
```

```python
fp_final.info()
```

```python
fp_final['Dept_hour'] = fp_final['Dep_Time'].str.split(":").str[0]
fp_final['Dept_min'] = fp_final['Dep_Time'].str.split(" ").str[1]
fp_final['Dept_hour'] = fp_final['Dept_hour'].astype('Int64')
fp_final['Dept_min']  = fp_final['Dept_min'].astype('Int64')
fp_final.drop('Dep_Time',axis= 1,inplace = True)
```
